```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)

df[["Order Date", "Ship Date"]].dtypes
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="mixed")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="mixed")

df["Lead Time"] = (df["Ship Date"] - df["Order Date"]).dt.days

df["Lead Time"].describe()
```

```python
print("Order Date range:")
print(df["Order Date"].min(), "to", df["Order Date"].max())

print("\nShip Date range:")
print(df["Ship Date"].min(), "to", df["Ship Date"].max())
```

```python
df["Lead Time Months"] = (
    (df["Ship Date"] - df["Order Date"]).dt.days / 30
).round(1)

df["Lead Time Months"].describe()
```

```python
product_factory_map = {
    "Wonka Bar - Nutty Crunch Surprise": "Lot's O' Nuts",
    "Wonka Bar - Fudge Mallows": "Lot's O' Nuts",
    "Wonka Bar -Scrumdiddlyumptious": "Lot's O' Nuts",
    "Wonka Bar - Milk Chocolate": "Wicked Choccy's",
    "Wonka Bar - Triple Dazzle Caramel": "Wicked Choccy's",
    "Laffy Taffy": "Sugar Shack",
    "SweeTARTS": "Sugar Shack",
    "Nerds": "Sugar Shack",
    "Fun Dip": "Sugar Shack",
    "Fizzy Lifting Drinks": "Sugar Shack",
    "Everlasting Gobstopper": "Secret Factory",
    "Hair Toffee": "The Other Factory",
    "Lickable Wallpaper": "Secret Factory",
    "Wonka Gum": "Secret Factory",
    "Kazookles": "The Other Factory"
}
```

```python
df["Factory"] = df["Product Name"].map(product_factory_map)

df[["Product Name", "Factory"]].head()
```

```python
df["Factory"].isnull().sum()
```

```python
factory_coords = {
    "Lot's O' Nuts": (32.881893, -111.768036),
    "Wicked Choccy's": (32.076176, -81.088371),
    "Sugar Shack": (48.11914, -96.18115),
    "Secret Factory": (41.446333, -90.565487),
    "The Other Factory": (35.1175, -89.971107)
}
```

```python
df["Factory_Lat"] = df["Factory"].map(lambda x: factory_coords[x][0])
df["Factory_Lon"] = df["Factory"].map(lambda x: factory_coords[x][1])

df[["Factory","Factory_Lat","Factory_Lon"]].head()
```

```python
region_coords = {
    "Interior": (39.7392, -104.9903),   # Denver
    "Atlantic": (36.8508, -76.2859),    # East coast
    "Gulf": (29.7604, -95.3698),        # Houston
    "Pacific": (34.0522, -118.2437)     # Los Angeles
}
```

```python
df["Customer_Lat"] = df["Region"].map(region_coords).str[0]
df["Customer_Lon"] = df["Region"].map(region_coords).str[1]
```

```python
df[df["Customer_Lat"].isnull()]["Region"].unique()
```

```python
from geopy.distance import geodesic

def calculate_distance(row):
    factory_loc = (row["Factory_Lat"], row["Factory_Lon"])
    customer_loc = (row["Customer_Lat"], row["Customer_Lon"])
    return geodesic(factory_loc, customer_loc).km

df["Distance_km"] = df.apply(calculate_distance, axis=1)
```

```python
df["Distance_km"].describe()
```

```python
df.to_csv("data/processed.csv", index=False)
```
